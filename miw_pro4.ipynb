```python
import numpy as np
import matplotlib.pyplot as plt


number_of_files = 15
for n in range(number_of_files):

  file = open("pro4_dane{}.txt".format(n + 1), "r")

  X = []
  y = []

  for f in file:
    d = f.split(" ")
    d.remove('/n')
    X.append(float(d[0]))
    y.append(float(d[1]))

  print(X)
  print(y)

  plt.plot(X, y, 'b*')
  plt.show()
```
